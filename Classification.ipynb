```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv('Chur.csv')
```

```python
df.head()
```

```python
df['Geography'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

```

```python
df['Geography']=encoder.fit_transform(df['Geography'])
df['Gender']=encoder.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
df.sample(7)
```

```python
df
```

```python
df=pd.get_dummies(df,columns=['Geography'],drop_first=True)
```

```python
df
```

```python
x=df.drop(columns=['Exited'],axis=1).values
```

```python
x
```

```python
y=df['Exited'].values
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python
x.shape
```

```python
model.add(Dense(11,activation='relu',input_dim=x_train.shape[1]))
model.add(Dense(11,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(x_train, y_train, epochs=150, batch_size=50, validation_split=0.2,verbose=1)
```

```python
import numpy as np

y_pred = np.argmax(model.predict(x_test), axis=-1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
